```python
import numpy as np
import os,sys
```

```python
from scipy.spatial.distance import pdist
```

```python


mode_settings_dict = {"0":"recon",
                      "1":"recon_class"}
```

```python
data_files =  [ _ for _ in os.listdir(os.getcwd()) if "data" in _]
```

```python
data_files
```

```python
data_files_ext = [_.split("_")[-1] for _ in data_files]
```

```python
dists = []
for file,file_ext in zip(data_files,data_files_ext):
    data = np.load(file)
    dist = pdist(data)
    dists.append(dist)
dists = np.array(dists)
```

```python
print(np.mean(dists[:2],axis=1))
print(np.mean(dists[2:],axis=1))
```
