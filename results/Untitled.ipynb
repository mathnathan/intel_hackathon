```python
import numpy as np
import os,sys
```

```python
from scipy.spatial.distance import pdist
from sklearn.manifold import TSNE
```

```python


mode_settings_dict = {"0":"recon",
                      "1":"recon_class"}


data_files =  np.array([ _ for _ in os.listdir(os.getcwd()) if "data" in _])
data_files_ext = np.array([_.split("_")[-1] for _ in data_files])


args = np.argsort(data_files_ext)
data_files = data_files[args]
data_files_ext = data_files_ext[args]


dists = []
for file,file_ext in zip(data_files,data_files_ext):
    data = np.load(file)
    dist = pdist(data)
    dists.append(dist)
dists = np.array(dists)
```
